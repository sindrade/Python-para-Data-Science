## Exercícios de estruturas Condicionais

```python
#1) Escreva um programa que peça à pessoa usuária para fornecer dois números e exibir o número maior.
num1 = float(input("Informe um número: "))
num2 = float(input("Informe outro núemro: "))

if num1 > num2:
    print(f"O número maior é: {num1}")
else:
    print(f"O número maior é: {num2}")
```

```python
# 2) Escreva um programa que solicite o percentual de crescimento de produção de uma empresa e informe se houve 
# um crescimento (porcentagem positiva) ou decrescimento (porcentagem negativa).

percentual = float(input("Qual o percentual de crescimento de produção da empresa? "))
if percentual > 0:
    print("\nHouve um crescimento na produção.")
else:
    print("\nHouve um descrescimento na produção.")
```

```python
# 3) Escreva um programa que determine se uma letra fornecida pela pessoa usuária é uma vogal ou consoante.
vogais = ["a", "e", "i", "o", "u"]
letra = input("Digite uma letra: ").upper()
if letra in vogais:
    print(f'A letra "{letra}", que foi digitada, é uma vogal!')
else:
    print(f'A letra "{letra}", que foi digitada, é uma consoante!')
```

```python
#4) Escreva um programa que leia valores médios de preços de um modelo de carro por 3 anos consecutivos e 
# exiba o valor mais alto e mais baixo entre esses três anos.
preco1 = 45001
preco2 = 45000
preco3 = 30000
if preco1 > preco2 > preco3:
    print(f"O preço mais alto é do ano 01: {preco1}")
elif preco2 > preco1 > preco3:
    print(f'O preço mais alto é do ano 02: {preco2}')
else:
    print(f'O preço mais alto é do ano 03: {preco3}')
    
if preco1 < preco2 < preco3:
    print(f'O preço mais baixo é do ano 01: {preco1}')
elif preco2 < preco1 < preco3:
    print(f'O preço mais baixo é do ano 02: {preco2}')
else:
    print(f'O preço mais baixo é do ano 03: {preco3}')
```

```python
#5) Escreva um programa que pergunte sobre o preço de três produtos e indique qual é o produto mais barato para comprar.
produto1 = float(input("Qual o preço do produto 01: "))
produto2 = float(input("Qual o preço do produto 02: "))
produto3 = float(input("Qual o preço do produto 03: "))

if produto1 < produto2 < produto3:
    print(f"\nO produto mais barato é o de preço: {produto1}")
elif produto2 < produto1 < produto3:
    print(f"\nO produto mais barato é o de preço: {produto2}")
else:
    print(f"\nO produto mais barato é o de preço: {produto3}")
```
